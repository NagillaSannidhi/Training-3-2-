```python
import re

text = input("Enter text: ")
date = re.search(r"\b\d{2}/\d{2}/\d{4}\b", text)
print(date.group() if date else "No date found")
```

```python
import re

text = input("Enter document text: ")

date = re.search(r"\b\d{2}/\d{2}/\d{4}\b", text)
amount = re.search(r"\d+", text)

print("Date:", date.group() if date else None)
print("Amount:", amount.group() if amount else None)
```

```python
import json

with open("document_data.json", "r") as f:
    data = json.load(f)

if data["amount"] > 40000:
    data["status"] = "HIGH VALUE"
else:
    data["status"] = "NORMAL"

with open("document_data.json", "w") as f:
    json.dump(data, f, indent=4)

print(data)
```
