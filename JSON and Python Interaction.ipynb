```python
import json

student = {
    "name": "Tarun",
    "roll_number": 101,
    "marks": 78
}

with open("student.json", "w") as file:
    json.dump(student, file, indent=4)
```

```python
with open("student.json", "r") as file:
    student = json.load(file)

print(student)
```

```python
name = student["name"]
roll = student["roll_number"]
marks = student["marks"]

print(f"Name: {name}")
print(f"Roll Number: {roll}")
print(f"Marks: {marks}")
```

```python
if marks >= 40:
    result = "Pass"
else:
    result = "Fail"

print("Result:", result)
```

```python
student["result"] = result

with open("student_result.json", "w") as file:
    json.dump(student, file, indent=4)

print("Updated JSON saved as student_result.json")
```
